```python
import glob
control_file=sorted(glob.glob("data/processed/control/*.csv"))
df1 = pd.DataFrame()
print(len(control_file))
for i in range(len(control_file)):
    df1=df1.append(pd.read_csv(control_file[i]))
df1['label']=0
print(df1.shape)
df1.head()
```

```python
park_file=sorted(glob.glob("data/processed/parkinson/*.csv"))
df2 = pd.DataFrame()
print(len(park_file))
for i in range(len(park_file)):
    df2=df2.append(pd.read_csv(park_file[i]))
df2['label']=1
print(df2.shape)
df2.head()
```

```python
df=pd.concat([df1, df2],axis=0)
print(df.shape)
df.head()
```

```python
def normalize(data):
    from sklearn.preprocessing import StandardScaler
    scaler = StandardScaler()
    output_columns = ["label"] 
    x=df.drop(output_columns,axis=1).columns.tolist()
    for feature in x:
        data[feature] = scaler.fit_transform(data[feature].values.reshape(-1,1))
    return data
df.head()
```

```python
df=normalize(df)
df.head()
```
